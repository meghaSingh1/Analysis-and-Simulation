```python
import matplotlib.pyplot as plt
import cmath
import numpy as np
```

```python
n= 1000
Bo= 1
gama = 0.5
```

```python
for r0 in np.arange(0.1, 5.0, 0.001):
    #beta =0.5
    #gama = beta/r0
    beta = r0*gama
    #print(beta,r0)
    
    #y = ((Bt*Bt)/(2*Bo))- ((beta*Bt)/(Bo*n))+((beta*Bt)/N)- ((beta+gama)/2)
    Bstar= Bo*((beta/n)*((1/Bo)+1)+cmath.sqrt(((beta/n)*((1/Bo)+1)+((beta+gama)/Bo))*((beta/n)*((1/Bo)+1)-((beta+gama)/Bo))))
    B2star= Bo*((beta/n)*((1/Bo)+1)-cmath.sqrt(((beta/n)*((1/Bo)+1)+((beta+gama)/Bo))*((beta/n)*((1/Bo)+1)-((beta+gama)/Bo))))
    #print(Bstar,B2star)
    plt.plot(r0,B2star,'g*')
    plt.plot(r0,Bstar, 'r*')
plt.title("equillibrim plot on varing value of r0")
plt.ylabel("B*")
plt.xlabel("R0")
plt.show()
```

```python
plt.figure()

for y in range(1, 100):
    r0 = 0.1
    beta = r0*gama
    Bt = ((y*y)/(2*Bo))- ((beta*y)/(Bo*n))+((beta*y)/n)- ((beta+gama)/2)
    plt.plot(y,Bt,'ro')
    plt.plot(r0,Bt,'go')
    r0 = r0 + 0.01
plt.title("Bifurcation plot")
plt.xlabel("Time")
plt.ylabel("Expression for [B](t)")
plt.show()
```

# Bifurcation Plot

```python
n= 1000
Bo= 1
gama = 0.5
```

```python
plt.figure()
for y in range(1, 100):
    r0 = 0.1
    beta = r0*gama
    Bt = ((y*y)/(2*Bo))- ((beta*y)/(Bo*n))+((beta*y)/n)- ((beta+gama)/2)
    plt.plot(y,Bt,'ro')
    #plt.plot(r0,Bt,'go')
    r0 = r0 + 0.01
plt.title("Bifurcation plot")
plt.xlabel("R0")
plt.ylabel("Expression for [B](t)")
plt.show()
```

```python
plt.figure()
for b in range(1,1000):
    Bt = ((beta*b*b)/2)- ((beta*b*b*b)/(n*3)) - ((gama*b*b)/2)
    plt.plot(b,Bt, 'bo')
plt.title("Bifurcation plot")
plt.xlabel("R0")
plt.ylabel("Expression for [B](t)")
plt.show()
```

```python
plt.figure()
for b in range(Bo,1000):
    Bt = ((beta*b*b)/2)- ((beta*b*b*b)/(n*3)) - ((gama*b*b)/2)
    plt.plot(b,Bt, 'bo')
plt.title("Bifurcation plot")
plt.xlabel("R0")
plt.ylabel("Expression for [B](t)")
plt.show()
```
